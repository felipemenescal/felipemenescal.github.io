# Apparent retrograde motion

```python
# import all libraries needed
from numpy import pi,cos,sin,arange,loadtxt,datetime64,timedelta64,arctan2
import pandas as pd
import matplotlib.pyplot as plt
```

## Introduction

Every two years, when observing Mars' position from Earth in relation to the celestial vault, the red planet appears to make a brief U-turn before continuing in the same direction. This retrograde motion has been observed since ancient times.

### Data

The data is extracted from [here](http://vo.imcce.fr/webservices/miriade/?forms).

The `DataMars.csv` file contains the coordinates (longitude and latitude) of Mars' position relative to the center of the Earth for dates between 2015 and 2017. Note that latitude is defined here relative to the plane of the ecliptic. The $x$ and $y$ axes are in the same direction for both the heliocentric (centered on the Sun) and geocentric (centered on the Earth) reference frames.

The center of the loop can be observed on May 25, 2016, at a longitude of 240.5°.

The apparent retrograde motion is observed when, in the heliocentric reference frame, the Earth “overtakes” Mars. At the “midpoint” of the retrograde motion, the Earth and Mars are therefore in the same direction relative to the Sun (the Sun, Earth, and Mars are aligned). By carefully choosing the date $t=0$ and the phase at the origin for both planets, we can make the model coincide with the data at that moment.

Useful data that we might need:

- Earth orbit radius : 1 AU;  
- Earth year: 365.256 d;
- Maris orbit radius: 1.52371 AU;  
- Mars year: 686.885 d;
- Original phase $\phi_0 = 240.5°$.

### Model

You will use a two-dimensional model of the motion of Earth and Mars around the Sun. The motion of both planets will be considered uniform circular and contained within the plane of the ecliptic.

This model will allow you to calculate the longitude of Mars as seen from Earth (i.e., the angle $\phi_\mathrm{apparent}$ in the geocentric reference frame) as a function of time.

We will use a graph to illustrate your model. To do so, we must represent both the data and the model on a graph.

Recalling: decimal degrees = degrees + (minutes/60) + (seconds/3600)

```python
# observational data

data = pd.read_csv('DataMars.csv',delimiter=',',header=None,comment = '#')

# show the table

data
```

### Parse the data

```python
long = pd.to_numeric(data.iloc[1:,3]) # slicing the longitude and converting to numerical information
date = pd.to_datetime(data.iloc[1:,1]) # converting the date info to datetime

date
```

### Utility functions

```python
# convert degree to rad
def rad(x): 
    return x * 2*pi/360

# convert rad to degree
def deg(x):
    return x * 360/(2*pi)
```

### Constants

```python
# parameters

TE = 365.256 # Earth year in days
TM = 686.885 # Mars year in days
rE = 1.0     # Earth orbit radius in AU
rM = 1.52371 # Mars orbit radius in AU
phi0 = 240.5 # initial phase in degree
```

### Physical equations

-  $x_\mathrm{planet}(t) = R\cos(\omega t+\phi_0)$ ($x$ position of the circular motion using the Sun frame)
-  $y_\mathrm{planet}(t) = R\sin(\omega t+\phi_0)$ ($y$ position of the circular motion using the Sun frame)
-  $x_\mathrm{TM}(t) = x_\mathrm{Mars} - x_\mathrm{Earth}$ ($x$ position of Mars using the Earth frame)
-  $y_\mathrm{TM}(t) = y_\mathrm{Mars} - y_\mathrm{Earth}$ ($y$ position of Mars using the Earth frame)
-  Longitude $= \arctan(y_\mathrm{TM}/x_\mathrm{TM})$ (longitude)
-  Angular velocity $\omega ={2\pi}/{T}$

### Picking the timeframe

As the center of the loop can be observed on May 25, 2016, at a longitude of 240.5°.

```python
time = arange(datetime64('2016-01-01'),datetime64('2016-10-26') + timedelta64(1, 'D'), dtype='datetime64[D]') # creates the datatime64 array

# simulation time frame
days = 150 # half amount of days
t = arange(-days,days) # modeling time, mirrored time frame at = 0, on May 25, 2016

# position listing
xearth = []
yearth = []
xmars = []
ymars = []

# calculation of longitudes
longitudes = []

for i in range(len(t)):
    xE = rE*cos(2*pi*t[i]/TE + rad(phi0))
    yE = rE*sin(2*pi*t[i]/TE + rad(phi0))
    
    xearth.append(xE)
    yearth.append(yE)

    xM = rM*cos(2*pi*t[i]/TM + rad(phi0))
    yM = rM*sin(2*pi*t[i]/TM + rad(phi0))
    
    xmars.append(xM)
    ymars.append(yM)
    
    # ME stands for Mars-Earth
    xME = xM - xE
    yME = yM - yE
    
    longitudes.append(deg(arctan2(yME,xME)) + 360) # takes the arctan then converts the longitude from rad to degrees and +360 corrects the angle value
```

Then we shall obtain this visualization of the orbits of Earth and Mars over these 300 days according to our model.

```python
plt.plot(xearth,yearth,label='Earth',color='b') # Earth
plt.plot(xmars,ymars,label='Mars',color='r') # Mars

# drawing the central positions for both Earth and Mars (t = 0)
plt.scatter(xearth[days],yearth[days],color='b',label='',s=10)
plt.scatter(xmars[days],ymars[days],color='r',label='',s=10)

plt.scatter(xearth[0],yearth[0],color='black',label='',s=5) # drawing Earth initial position at t[0] = -150
plt.scatter(xmars[0],ymars[0],color='black',label='',s=5) # drawing Mars initial position at t[0] = -150

plt.scatter(0,0,color='y',label='Sun',s=200) # drawing a single yellow dot at the center for the Sun

plt.gca().set_aspect('equal') # set plot equal aspect ratio

plt.xlabel("$x$ (AU)")
plt.ylabel("$y$ (AU)")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.title('Earth and Mars positions over ' + str(2*days) + ' days')
plt.legend()
plt.grid()
plt.show()
```

## Poltting Earth's POV (model)

```python
plt.plot(t,longitudes, label="model", color='b')
plt.xlabel("Time (days) centered at May 25, 2016")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Longitude (°)")
plt.ylim(190,290)
plt.title("Mars longitude from Earth frame")
plt.legend()
plt.show()
```

## Poltting Earth's POV (model *versus* observation)

```python
# set date lims
left = pd.to_datetime('2016-01-01')
right = pd.to_datetime('2016-10-10')
center = pd.to_datetime('2016-05-25')

dates = []

# background stars generator

import random

rndlong = []
rnddate = []

rnd = 2000 # amount of stars

for i in range(rnd):
    rndlong.append(random.uniform(190,290))
    rnddate.append(random.randint(-days,days))

# date list from modeling

for k in range(len(t)):
    dates.append(center + timedelta64(t[k], 'D'))
    # random dates
    rnddate[k] = center + timedelta64(rnddate[k], 'D')

# creating the back ground

plt.scatter(rnddate,rndlong,color='white',marker='v',s=0.1)

# plotting the model + data

plt.plot(dates,longitudes,label="model", color='white')
plt.plot(date,long,label="observation", color='yellow')
plt.gca().set_facecolor('black')
plt.xlabel("Date")
plt.xticks(rotation=45,ha='right')
plt.ylabel("Longitude (°)")
plt.xlim(left,right)
plt.ylim(190,290)
plt.title("Mars longitude from Earth frame")
plt.legend(loc='upper left',labelcolor='white')
plt.show()
```
